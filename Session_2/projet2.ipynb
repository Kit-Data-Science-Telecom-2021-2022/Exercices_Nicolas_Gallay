**Projet maison n° 1**

- Ajoutez une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.
- Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.
- Ajoutez une fonction de conversion pour pouvoir utiliser la première fonction avec un GPS (degrés, minutes, secondes).

```python
#Ajoutez une colonne 'CP Ville' avec le Code postal + un espace + et le nom de la Ville.

geo['CP Ville'] = geo['Code Postal'].str.cat(geo['Commune'], sep=" ")
geo['CP Ville']
```

```python
# Ecrivez une fonction qui détermine la commune la plus proche d'un point à partir de sa latitude et sa longitude.
# fonction recherche de ville
def ville(lat, long):
    geo['distance_carree'] = (geo['Latitude'] - lat)**2 + (geo['Longitude'] - long)**2
    commune_proche = geo.loc[geo['distance_carree'].idxmin(), ['Commune']]
    return commune_proche 
```

```python
ville(44.32,3.5)
```

```python
# conversion degrés, minutes, secondes => décimal
def dms2dec(deg, mn, sec):
    return deg + mn / 60 + sec / 3600
```

```python
# on applique la fonction à une coordonnée tirée au hasard
# init du random
np.random.seed(0)

a, b = 41.5, 51.1  # latitude min et max de la France métropolitaine
lat = (b - a) * np.random.random() + a
a, b = -5.1, 9.5  # longitude min et max de la France métropolitaine
long = (b - a) * np.random.random() + a
ville(lat, long)

```

```python
# à partir de coordonnées GPS précises
ville(dms2dec(48, 42, 52), dms2dec(2, 14, 45))
```

Sur Windows shapely peut nécessiter d'installer la dll "geos_c.dll" dans le répertoire "Library/bin" de votre environnement Python.

```python
#Utiliser la librairie Python json pour parser la colonne "geo_shape".
geo['geo_shape']
```

```python
geo = geo.reset_index()
df_json = pd.json_normalize(geo['geo_shape'].apply(json.loads))
geo = geo.assign(type = df_json['type'], coordinates = df_json['coordinates'])
```

```python
geo = geo.set_index('Code INSEE')
```

```python
geo[['type','coordinates']].head()
```

```python
#Donner le décompte des valeurs accédées avec la clé "type".
geo['type'].value_counts()
```

```python
#Donner le décompte des longueurs des listes accédées avec la clé "coordinates".
geo['coordinates'].str.len().value_counts()
```

```python
#Quelle commune est la plus complexe géométriquement ?
#Celle avec le plus de liste dans coordinates
geo.loc[geo['coordinates'].str.len().idxmax(),'Commune']
```

```python
geo['coordinates'].str.len().idxmax()
```

```python
#Quelle commune est la seconde la plus complexe géométriquement ?
geo.loc[geo['coordinates'].str.len().sort_values(ascending=False).index[1]]['Commune']
```

```python
#Quelles sont les villes qui sont de type "Polygon" mais dont la longueur des listes accédées avec la clé "coordinates" vaut 2 ?
geo.loc[(geo['coordinates'].str.len() == 2) & geo['type'].str.fullmatch('Polygon')]['Commune']
```

```python
#Pour ces villes vérifier que le premier polygone contient bien le second (enclave).

polygons_serie = geo.loc[(geo['coordinates'].str.len()==2) & geo['type'].str.fullmatch('Polygon')]

figure = plt.figure(figsize=(10, 20), dpi=100)

for index, polygon_liste in enumerate(polygons_serie['coordinates']):
    plt.subplot(5, 2, index+1)
    
    for polygon in polygon_liste:
        x, y=Polygon(polygon).exterior.xy
        plt.plot(x, y)
        
    plt.title(geo['Commune'].loc[polygons_serie.index[index]])
        
plt.show()
```
